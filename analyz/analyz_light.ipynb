```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('data/data-61762-2025-09-25.csv', sep=';')
df.head()
```

```python
# снесем первую строку
df = df.drop(index=0).reset_index(drop=True)
df.head()
```

```python
df['geoData'] = df['geoData'].str.replace('{', '')
df['geoData'] = df['geoData'].str.replace('}', '')
df['geoData'] = df['geoData'].str.replace('], type=Point', '')
df['geoData'] = df['geoData'].str.replace('coordinates=[', '')


df['geoData'].iloc[0]
```

```python
df[['x', 'y']] = df['geoData'].str.split(', ', expand=True)
df.head()
```

```python
gdf = gpd.GeoDataFrame(data=df, geometry=gpd.points_from_xy(x=df['x'], y=df['y']), crs=4326)
gdf.head()
```

```python
gdf.to_file('data/light.geojson')
```

```python
# считаем буфферы в utm зоне
gdf_32637 = gdf.to_crs(32637).buffer(10)
```

```python
# переводим в веб-стандарт проекцию
gdf_3857 = gdf_32637.to_crs(3857)
```

```python
gdf_3857.to_file('data/buffer.gpkg')
```
