Анализ данных по дтп

```python
import pandas as pd
```

```python
df = pd.read_csv('data/accidents_all_regions_126_v20240621.csv', sep=';')
df.head()
```

```python
import geopandas as gpd

gdf = gpd.GeoDataFrame(data=df, geometry=gpd.points_from_xy(x=df['longitude'], y=df['latitude']), crs=4326)
gdf.head()
```

```python
gdf_mos = gdf[(gdf['region'] == 'Москва') | (gdf['region'] == 'Московская область')]
gdf_mos.head()
```

```python
gdf_mos.to_file('data/dtp_mosreg.gpkg')

# в дальнейшем необходима обработка через gis для удаления выбросов
```
