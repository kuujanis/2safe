Анализ датасета с камерами по Москве

```python
import pandas as pd
import geopandas as gpd
```

```python
# загрузка данных
df = pd.read_csv('data/data-1498-2024-11-25.csv', sep=';')
df.head()
```

```python
# снесем первую строку
df = df.drop(index=0).reset_index(drop=True)
df.head()
```

```python
df['geoData'] = df['geoData'].str.replace('{', '')
df['geoData'] = df['geoData'].str.replace('}', '')
df['geoData'] = df['geoData'].str.replace('], type=Point', '')
df['geoData'] = df['geoData'].str.replace('coordinates=[', '')


df['geoData'].iloc[0]
```

```python
df[['x', 'y']] = df['geoData'].str.split(', ', expand=True)
df.head()
```

```python
gdf = gpd.GeoDataFrame(data=df, geometry=gpd.points_from_xy(x=df['x'], y=df['y']), crs=4326)
gdf.head()
```

```python
gdf.to_file('data/cameras.gpkg')
```
