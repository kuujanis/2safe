```python
import geopandas as gpd
import matplotlib.pyplot as plt
import momepy
import networkx as nx
```

```python
import momepy
gdf_open_mos = gpd.read_file('data/routes_mos_open.gpkg')
gdf_open_mos = gdf_open_mos.to_crs(3857)
```

```python
gdf_open_mos = gdf_open_mos.explode(ignore_index=True)
```

```python
# 2. Создаем граф
# mode='undirected' — дороги без направления
G = momepy.gdf_to_nx(gdf_open_mos, approach='primal')
```
